```python
%%sql
create or REPLACE TABLE LH_Refined.bad_data(
    RunID VARCHAR(400),
    Execution_DateTime TIMESTAMP,
    Entity_Name VARCHAR(400),
    Entity_Key VARCHAR(400),
    Reference_Key VARCHAR(400),
    Reason string,
    Table_Data string
) USING DELTA;
```

```python
%%sql
CREATE or REPLACE TABLE LH_Refined.activity_log
(
    RunID VARCHAR(400),
    Source_Location varchar(200),
    Source_Type varchar(200),
    Source_Name varchar(200),
    Destination_Location varchar(200),
    Destination_Type varchar(200),
    Destination_Name varchar(200),
    Start_Time TIMESTAMP,
    End_Time TIMESTAMP,
    Activity varchar(200),
    Source_Count long,
    Destination_Count long,
    BadData_Count long

) USING DELTA;
```

```python
%%sql
CREATE or replace TABLE LH_Refined.logdata
(
    RunID varchar(400),
    SequenceNo long,
    LogDate TIMESTAMP,
    LogType varchar(400),
    ObjectName varchar(400),
    ActivityName varchar(400),
    Description string

) USING DELTA;
```
